```python
#Univariate analysis
def univariate_analysis(df):
    for col in df.columns:
        print(f"Univariate Analysis for '{col}':")
        if df[col].dtype == 'object':
            
            plt.figure(figsize=(8, 4))
            sns.countplot(data=df, x=col, palette='Set3')
            plt.title(f'Countplot of {col}')
            plt.xticks(rotation=45)
            plt.show()
        elif df[col].dtype in ['int64', 'float64']:
            
            plt.figure(figsize=(12, 4))
            plt.subplot(1, 2, 1)
            sns.histplot(df[col], bins=20, kde=True)
            plt.title(f'Visualizing the Distribution of {col}')
            plt.subplot(1, 2, 2)
            sns.boxplot(data=df, y=col, palette='Set2')
            plt.title(f'Boxplot of {col}')
            plt.tight_layout()
            plt.show()


#Bivariate Analysis
def clustering_corr_matrix(df):
    corr_matrix = df.corr()

    num_cols = df.shape[1]
    placeholder_cols = [f"Column {i+1}" for i in range(num_cols)]
    corr_matrix_with_placeholders = pd.DataFrame(corr_matrix.values, columns=placeholder_cols, index=placeholder_cols)

    plt.figure(figsize=(24, 20))
    sns.set(style="white")
    sns.clustermap(corr_matrix_with_placeholders, cmap="coolwarm", annot=False, linewidths=0.5)

    legend_labels = [f"{col}: {placeholder_cols[i]}" for i, col in enumerate(df.columns)]
    ax = plt.gca()
    ax.text(22, 0.0005, '\n'.join(legend_labels), transform=ax.transAxes, fontsize=12, va="center", ha="left")

    legend = ax.text(22, 1.9, "Legend:", transform=ax.transAxes, fontsize=20, va="center", ha="left", weight="bold")
    legend.set_bbox(dict(facecolor='white', alpha=0.6))

    plt.title("Correlation Heatmap with Clustered Rows and Columns", fontsize= 25, y=1.9, fontweight="bold")
    plt.show()
    
#Missing values


def handleMissing_values(df):
    missing_values = df.isnull().sum()
    missing_columns = missing_values[missing_values > 0]

    if missing_columns.empty:
        # If there are no missing values, display a message
        print("No missing values in the DataFrame.")
    else:
        # Percentage of missing values
        missing_percentage = round((missing_columns / len(df)) * 100, 2)

        # Putting the missing count and also the percentage into a single DataFrame for easy viewing of the table at the end
        missing_info = pd.DataFrame({'Count of Missing Values': missing_columns, 'Percentage of missing values (%)': missing_percentage})

        # Sorting the DataFrame by the number of missing values in descending order
        missing_info_sorted = missing_info.sort_values(by='Count of Missing Values', ascending=False)

        fig, ax = plt.subplots(figsize=(12, 8))

        ax.axis('off')

        missing_info_combined = pd.DataFrame({'Column Names': missing_info_sorted.index,
                                            'Count of Missing Values': missing_info_sorted['Count of Missing Values'],
                                            'Percentage of missing values (%)': missing_info_sorted['Percentage of missing values (%)']})

        table = ax.table(cellText=missing_info_combined.values, colLabels=missing_info_combined.columns, cellLoc='center', loc='center', colColours=['#f2f2f2'] * (len(missing_info_combined.columns)))
        table.auto_set_font_size(False)
        table.set_fontsize(12)
        plt.show()



#Standardize
def standardize_data(df): 
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    df[numeric_cols] = StandardScaler().fit_transform(df[numeric_cols])

    return df

#PCA

def pca_implementation(df):
    df= df.dropna()
    pca = PCA(n_components=2)

    principal_components = pca.fit_transform(df)

    principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

    explained_variance_ratio = pca.explained_variance_ratio_
    print("Explained Variance Ratios:", explained_variance_ratio)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(df)
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])


def visualize_principal_comp(df):
    plt.figure(figsize=(24, 10))
    plt.scatter(principal_df['PC1'], principal_df['PC2'], c='blue', alpha=0.5)
    plt.title('Visualizing the Principal Components', fontweight='bold', fontsize=25)
    plt.xlabel('Principal Component 1', fontsize=18)
    plt.ylabel('Principal Component 2', fontsize=18)
    plt.grid(False)
    plt.show()

#MODELLING
def choosing_k():
    inertia = []#In this list I will store the inertia (within-cluster sum of squares) values

    #This is the range of values for k (number of clusters)
    k_range = range(1, 11)

    #inertia for different values of k
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(principal_df)  # Assuming you've already performed PCA
        inertia.append(kmeans.inertia_)

    #elbow plot to visualize for the inertia values
    plt.figure(figsize=(8, 6))
    plt.plot(k_range, inertia, marker='o', linestyle='-', color='b')
    plt.title('Elbow Plot for K-Means Clustering')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()

def training_the_model():
    #Train K-Means with 3 as the number of clusters (k=3)
    kmeans = KMeans(n_clusters=3, random_state=42)
    cluster_labels = kmeans.fit_predict(principal_df.values)  # Use .values to access NumPy array

    #The cluster labels to act as a new column in principal_df
    principal_df['Cluster'] = cluster_labels

    # Visualizing the clusters
    plt.figure(figsize=(24, 10))
    plt.scatter(principal_df.iloc[:, 0], principal_df.iloc[:, 1], c=cluster_labels, cmap='viridis')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
    plt.title('K-Means Clustering (k=3)')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend()
    plt.show()
    return cluster_labels
def silhouette_plot():
    # Silhouette scores for each data points
    silhouette_values = silhouette_samples(principal_df, cluster_labels)

    plt.figure(figsize=(10, 6))
    y_lower = 10


    for i in range(3):# since k=3
        # This gives silhouette scores for the current cluster
        cluster_silhouette_values = silhouette_values[cluster_labels == i]
        cluster_silhouette_values.sort()

        #Size of current cluster
        cluster_size = cluster_silhouette_values.shape[0]

        #y-coordinate
        y_upper = y_lower + cluster_size

        #Coloring the current cluster
        color = plt.cm.viridis(float(i) / 3)
        plt.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_silhouette_values,
                        facecolor=color, edgecolor=color, alpha=0.7)

        # This Labels the silhouette plot with cluster numbers
        plt.text(-0.05, y_lower + 0.5 * cluster_size, str(i))

        y_lower = y_upper + 10
    plt.xlabel("Silhouette Score")
    plt.ylabel("Cluster")
    plt.title("Silhouette Plot for Each Cluster")

    #Average silhouette score
    average_silhouette_score = silhouette_score(principal_df, cluster_labels)
    plt.axvline(x=average_silhouette_score, color="red", linestyle="--", label="Average Silhouette Score")
    plt.legend()
    plt.show()

def silhouette_elbow_for_evaluation():
    silhouette_scores = []

    #k values (number of clusters) that I want to check 
    k_values = range(2, 11)

    # Silhouette scores for the different values of k
    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=42)
        cluster_labels = kmeans.fit_predict(principal_df.values)
        silhouette_avg = silhouette_score(principal_df.values, cluster_labels)
        silhouette_scores.append(silhouette_avg)

    plt.figure(figsize=(10, 5))
    plt.plot(k_values, silhouette_scores, marker='o', linestyle='-')
    plt.title('Silhouette Score vs. Number of Clusters')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Silhouette Score')
    plt.grid(True)
    plt.show()
```

## __4.2 Bivariate Analysis__

#### __For this analysis, I choose to make use of Hierarchical clustering to make a clustered correlation heatmap. This is a method for grouping similar data points into clusters, and here it represents its results with a correlation heatmap. The heatmap itself reflects the similarity in correlation patterns between variables.__ 

#### Blocks of dark (or light) cells indicate groups of variables that are highly positively (or negatively) correlated. Gaps or transitions in color intensity may represent the boundaries between different clusters of variables. Columns that are closer to each other in the heatmap tend to have more similar correlation patterns, as reflected in the dendrogram.
__<u>The process is as follows:</u>__

<li>A similarity measure (Pearson correlation coefficient) is used to quantify how similar or dissimilar columns are in terms of their correlation patterns.

<li>A distance matrix is calculated, that contains pairwise distances between all columns based on the chosen similarity measure. Highly correlated columns have a small distance, while uncorrelated columns have a large distance.

<li>A dendrogram, which is a hierarchical tree-like structure, is used to visualize the relationships between columns. The 'leaves' of this structure represent individual columns, while internal nodes represent clusters of columns, and their height indicates the merging distance.

<li>As can be seen in clustered correlation heatmap, rows and columns are reordered based on the hierarchical clustering dendrograms, grouping similar variables together. The dendrograms are displayed on the sides of the heatmap, showing how clusters of columns were merged at different levels of similarity.

Equation for Clustering:$$d(A, B) = \frac{1}{|A| * |B|} \sum_{i \in A} \sum_{j \in B} d(i, j)$$

__<u>In this equation:</u>__
d(A, B) represents the distance between two clusters A and B.
|A| and |B| represent the number of elements in clusters A and B, respectively.
The double summation calculates the sum of pairwise distances between all pairs of elements, where i is in cluster A and j is in cluster B.

The Elbow Method is used to determine the optimal number of clusters (k) in a clustering algorithm,and in this project,the K-Means. It involves plotting the within-cluster sum of squares (inertia) for a range of values of

LaTeX Equation:$$\text{W}(k) = \sum_{i=1}^{k} \sum_{j=1}^{n_k} ||x_j - \mu_i||^2$$

## __7.2 Model Training__

K-Means is a popular clustering algorithm used for partitioning a dataset into 'k' distinct, non-overlapping subgroups or clusters. It is an iterative algorithm that seeks to find the centroids (centers) of these clusters by minimizing the sum of squared distances between data points and their respective cluster centroids.

#### <u> The algorithm works as follows:</u>
 Initialize 'k' centroids randomly, where 'k' is the number of desired clusters.
 Assign each data point to the nearest centroid based on the Euclidean distance between the data point and the centroids.
 Recalculate the centroids as the mean of all data points assigned to each cluster.
 Repeat steps 2 and 3 until convergence, which occurs when the centroids no longer change significantly or a predefined number of iterations is reached.
 The formula to calculate the Euclidean distance between two data points, 'x' and 'c' (centroid), in a multi-dimensional space is given by: $$\text{Euclidean Distance}(x, c) = \sqrt{\sum_{i=1}^{n} (x_i - c_i)^2}$$

Here, 'n' represents the number of features (dimensions) in the data points.

In summary, K-Means is an unsupervised machine learning algorithm that divides a dataset into 'k' clusters by minimizing the distance between data points and cluster centroids. It's commonly used for tasks like customer segmentation, image compression, and more.

##          __7.3.1 Silhouette Score__

Silhouette Score: The Silhouette Score measures how similar an object is to its own cluster (cohesion) compared to other clusters (separation). It ranges from -1 to 1, where a higher value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters.

Formula for Silhouette Score:$$\text{Silhouette Score}(x) = \frac{b(x) - a(x)}{\max(a(x), b(x))}$$

In this formula:

a(x) represents the average distance from the data point x to other data points within the same cluster.
b(x) represents the minimum average distance from the data point x to data points in a different cluster, minimized over clusters.

```python
cluster_labels = training_the_model()
# Compute the Silhouette Score
silhouette_avg = silhouette_score(principal_df.values, cluster_labels)
print(f"Silhouette Score: {silhouette_avg}")
```

## __7.3.2 Inertia__

Description:
Inertia is a measure of the within-cluster sum of squares. It quantifies how far the points within a cluster are from the centroid of that cluster. Lower inertia values indicate that the data points in a cluster are closer to the centroid, which is desirable.

Equation (Within-Cluster Sum of Squares):
The inertia for a K-Means cluster with 

k clusters can be calculated as follows: $$\text{Inertia} = \sum_{i=1}^{k} \sum_{x \in C_i} \left\|x - \mu_i\right\|^2$$

Where:

"k" is the number of clusters.
"Ci" represents the i-th cluster.
"x" is a data point in cluster "Ci".
"i" is the centroid of cluster "Ci".
"x - i" represents the Euclidean distance between data point "x" and centroid "i".




```python
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(principal_df.values)
inertia = kmeans.inertia_
print(f"Inertia: {inertia}")
```

## __7.3.3 Davies-Bouldin Index__

The Davies-Bouldin Index is a metric used for evaluating the performance of clustering algorithms like K-Means. It measures the average similarity between each cluster and its most similar cluster. The lower the Davies-Bouldin Index, the better the clustering results.

The formula for the Davies-Bouldin Index can be represented in LaTeX as: $$\text{Davies-Bouldin Index} = \frac{1}{k} \sum_{i=1}^{k} \max_{j \neq i} \left(\frac{\sigma_i + \sigma_j}{d_{ij}}\right)$$

Where:

"k" is the number of clusters.
"i" is the average distance from each point in cluster "i" to the centroid of that cluster.
"dij" is the distance between the centroids of cluster "i" and cluster "j."
The Davies-Bouldin Index quantifies the average "cluster separation" across all clusters, providing a measure of how distinct and well-separated the clusters are from each other. Lower values indicate better-defined clusters.

```python
db_index = davies_bouldin_score(principal_df.values, cluster_labels)
print(f"Davies-Bouldin Index: {db_index}")
```

### __7.3.4 Calinski-Harabasz Index__

The Calinski-Harabasz Index is a metric used to evaluate the quality of clusters produced by a clustering algorithm, such as K-Means. It measures the ratio of between-cluster variance to within-cluster variance. A higher Calinski-Harabasz Index suggests better-defined and more separated clusters, indicating a better clustering solution.

The formula for the index is $$\text{Calinski-Harabasz Index} = \frac{B(k-1)}{W(k-1)}$$

Where:

B is the between-cluster sum of squares.
W is the within-cluster sum of squares.
k is the number of clusters.


```python
ch_index = calinski_harabasz_score(principal_df.values, cluster_labels)
print(f"Calinski-Harabasz Index: {ch_index}")
```
